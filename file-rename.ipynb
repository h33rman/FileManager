```python
import os

def rename_files_in_folder(folder_path, year):
    files = os.listdir(folder_path)
    for file_name in files:
        if file_name.endswith('.html'):
            # Extract the number from the file name and remove the last four digits if they match the year
            number = ''.join(filter(str.isdigit, file_name))
            if number.endswith(year):
                number = number[:-len(year)]
            # Create the new file name
            new_file_name = f"{year}-arrest-{number.zfill(3)}.html"
            # Get the full paths
            old_file_path = os.path.join(folder_path, file_name)
            new_file_path = os.path.join(folder_path, new_file_name)
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f"Renamed: {old_file_path} to {new_file_path}")
```

```python
# Define the main folder path
main_folder_path = '/home/herman/Downloads/Documents/Arrest/arrest_kij'

# List of years to process
years = ['2016','2017', '2018']
```

```python
# Process each year folder
for year in years:
    folder_path = os.path.join(main_folder_path, year)
    rename_files_in_folder(folder_path, year)
```
